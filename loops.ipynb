```python
#using nested loop
for i in range(1,6,1):
  for j in range(i):
    print("*",end=" ")
  print()
```

```python
for i in range(1,6):
  for j in range(i):
    print(j+1,end=" ")
  print()
```

```python
for i in range(5,0,-1):
  for j in range(i,0,-1):
    print(j,end=" ")
  print()
```
