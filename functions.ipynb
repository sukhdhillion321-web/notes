```python
#positional
def student(name,age):
  print("name",name)
  print("age",age)
student("sukh",19)
```

```python
#keyword
def student(name,age):
  print("name",name)
  print("age",age)
student(age=19,name="sukh")
```

```python
# *args
def fruits(*args):
  print(args)
fruits("apple","banana","mango","grapes")
```

```python
def fruits(*args):
  for i in args:
    print(i)
fruits("apple","banana","mango","grapes")
```

```python
# Variable-Length Keyword Parameters (**kwargs)
def student(**kwargs):
    print(kwargs)

student(name="PQR", age=20, course="Python", domain = "AI")
```

```python
def multi(list):
  for i in range(0,4):
    print(list[i]*list[i+1])
multi([1,3,4,5,6])
```

```python
def reverse(name):
  for i in range(-1,-6,-1):
    print(name[i],end=" ")
reverse("sukhi")
```

```python
def table(num):
  for i in range(1,11):
    print(num,"*",i,"=",num*i)
table(4)
```
