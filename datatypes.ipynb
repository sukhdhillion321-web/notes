```python
text = "sukhjit kaur dhillon"

print("Original :", text)
print("upper() :", text.upper())
print("lower() :", text.lower())
print("title() :", text.title())
print("capitalize() :", text.capitalize())
print("swapcase() :", text.swapcase())
print("casefold() :", text.casefold())
```

```python
text = "    sukhjit kaur dhillon    "

print("Original :", text)
print("strip() :", text.strip())
print("lstrip() :", text.lstrip())
print("rstrip() :", text.rstrip())
```

```python
text = "sukhjit kaur sukhjit"

print("find() :", text.find("sukhjit"))
print("rfind() :", text.rfind("sukhjit"))
print("index() :", text.index("kaur"))
print("rindex() :", text.rindex("sukhjit"))
print("count() :", text.count("sukhjit"))
```
