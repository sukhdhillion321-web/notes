```python
#arithemetic operator
a=input("enter the value of a"))
b=int(input("enter the value of b"))
print("sum:",a+b)
print("substract:",a-b)
print("multiply:",a*b)
print("divide:",a/b)
print("floor division:",a//b)
print("modolus:",a%b)
print("power:",a**b)
```

```python
#relational operator
a=int(input("enter the value of a"))
b=int(input("enter the value of b"))
print("comparisson:",a==b)
print("not equal to:",a!=b)
print("less than:",a<b)
print("greater than",a>b)
print("less than equal to:",a<=b)
print("greater than equal to:",a>=b)
```

```python
#assignment operator
a=int(input("enter the value of a"))
a+=5
print(a)
a-=3
print(a)
a*=2
print(a)
a/=2
print(a)
a%=2
print(a)
```

```python
#logical operator
age=int(input("enter the value of a"))

print(age > 18 and age < 25)

print(age > 18 or age > 50)

print(not(age > 18))
```

```python
#membership operator
colors="red","yellow","green"
print("red" in colors)
print("yellow" not in colors)
```

```python
#logical operator
a=[20,30]
c=[20,30]
b=a
d=c
print(a is c)
print(a is b)
print(b is a)
print(c is b)
print(c is d)
print(d is a)
print(a is not b)
```

```python
a = int(input("enter the value of a"))
b = int(input("enter the value of b"))

print("a =", a)
print("b =", b)

print("\nAND (&):", a & b)

print("OR (|):", a | b)

print("XOR (^):", a ^ b)

print("NOT (~a):", ~a)

print("Left Shift (a << 2):", a << 2)

print("Right Shift (a >> 1):", a >> 2)
```
